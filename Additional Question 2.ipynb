```python
from skimage.color import rgb2grey
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
from skimage.data import imread
```

```python
def convert(file):
    image = imread(file)
    image = rgb2grey(image)
    image = np.subtract(np.ones((8,8)),image)
    image = image*(16/image.max())
    image = np.array(image,dtype=int)
    return image
```

```python
plt.matshow(convert('E5.png'))
```

```python
E1 = convert('E1.png')
E2 = convert('E2.png')
E3 = convert('E3.png')
E4 = convert('E4.png')
E5 = convert('E5.png')
E6 = convert('E6.png')
E7 = convert('E7.png')
FS1 = convert('FS1.png')
FS2 = convert('FS2.png')
FS3 = convert('FS3.png')
FS4 = convert('FS4.png')
```

```python
training_set_images = [E1,E4,E6,FS1,FS3]
training_set_targets = ['Elliptical','Elliptical','Elliptical','Spiral','Spiral']
test_set_images = ['E2.png','E3.png','E5.png','E7.png','FS2.png','FS4.png']

test_set_targets = ['Elliptical','Elliptical','Elliptical','Elliptical','Spiral','Spiral']
```

```python

"""use training set to find average array for each number"""
def average_array(n): #n = 'Elliptical' or 'Spiral'
    count=0
    narray=np.zeros((8,8))
    for i in range(0,5):
        if training_set_targets[i] == n:
            count+=1
            narray+=training_set_images[i]
    return narray/count

"""name and save each array"""
Elliptical = average_array('Elliptical')
Spiral = average_array('Spiral')
```

```python
def guess_type(file):
    image = imread(file)
    image = rgb2grey(image)
    image = np.subtract(np.ones((8,8)),image)
    image = image*(16/image.max())
    image = np.array(image,dtype=int)
    a = [Elliptical,Spiral]
    guess = {n:np.sum(abs(np.subtract(a[n],image))) for n in range(2)}
    if min(guess, key=guess.get) == 0:
        return 'Elliptical'
    if min(guess, key=guess.get) == 1:
        return 'Spiral'
```
